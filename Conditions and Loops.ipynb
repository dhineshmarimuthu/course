# Python Conditions

Python supports the usual logical conditions from mathematics

Equals: a == b

Not Equals: a != b

Less than: a < b

Less than or equal to: a <= b

Greater than: a > b

Greater than or equal to: a >= b

These conditions can be used in several ways, most commonly in "if statements" and loops

An "if statement" is written by using the if keyword

```python
# Python relies on indentation (whitespace at the beginning of a line) to define scope in the code. 

# Other programming languages often use curly-brackets for this purpose

num1=10
num2=5
if num1>num2:
    print("num1 is largest")
```

```python
# The elif keyword is pythons way of saying "if the previous conditions were not true, then try this condition

num1=5
num2=5
if num1>num2:
    print("num1 is largest")
elif(num1==num2):
    print("num1 equals num2")
```

```python
# The else keyword catches anything which isn't caught by the preceding conditions

num1=3
num2=5
if num1>num2:
    print("num1 is largest")
elif(num1==num2):
    print("num1 equals num2")
else:
    print("num2 is largest")
```

```python
# The and keyword is a logical operator, and is used to combine conditional statements

num1=10
num2=7
num3=5
if num1>num2 and num2>num3:
    print("Passed")
else: 
    print("Failed")
```

```python
num1=10
num2=7
num3=5
if num1>num2 or num2>num3:
    print("Passed")
else: 
    print("Failed")
```

```python
# You can have if statements inside if statements, this is called nested if statements

mark1=0

if mark1>=0:
    if(mark1>70):
        print("First class")
    elif(50<mark1<=70):
        print("Second class")
    else:
        print("Third class")
else:
    print("Enter valid mark")
```

# Loops 

Python has two primitive loop commands:

while loops

for loops

### The while Loop

```python
# With the while loop we can execute a set of statements as long as a condition is 

# In while loop, we need to declare the increment or decrement variable before initiating the loop 

x=1

while x<=5:
    print("Exexuting loop number",x)
    x+=1
```

```python
# With the break statement we can stop the loop even if the while condition is true

x=1
while x<=5:
    print("Exexuting loop number",x)
    x+=1
    if(x==3):
        break
```

```python
# With the continue statement we can stop the current iteration, and continue with the next

x=0
while x<=5:
    x+=1
    if(x==3):
        continue
    print("Exexuting loop number",x)
    
```

```python
# With the else statement we can run a block of code once when the condition no longer is true

x=1

while x<=5:
    print("Exexuting loop number",x)
    x+=1
else:
    print("Loop completed")
```

### For loops 

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string)

With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.



```python
# The range() function defaults to 0 as a starting value, 
# however it is possible to specify the starting value by adding a parameter

# range(2, 6), which means values from 2 to 6 (but not including 6)

for i in range(3,10):
    print(i)
```

```python
# The range() function defaults to increment the sequence by 1, 
# however it is possible to specify the increment value by adding a third parameter

for i in range(3,10,3):
    print(i)
```

### Nested Loops

A nested loop is a loop inside a loop.

The "inner loop" will be executed one time for each iteration of the "outer loop"

```python
for i in range(1,4):
    for j in range(2,7,2):
        print(i,"+",j,"=",i+j)
```

```python
# for loops cannot be empty, but if you for some reason have a for loop with no content, 
# put in the pass statement to avoid getting an error

for i in  range (1,6):
    if(i==3):
        pass
    else:
        print(i)
```
